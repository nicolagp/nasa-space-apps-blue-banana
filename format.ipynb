```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = "https://opendata.ecdc.europa.eu/covid19/casedistribution/csv"
df = pd.read_csv(url)
df.head()
```

```python
df["dateRep"] = pd.to_datetime(df["dateRep"])
df = df.set_index(df["dateRep"]).drop("dateRep", axis=1)
df = df.sort_index()
```

```python
plt.plot(df[df["geoId"] == "US"]["deaths"])
```
